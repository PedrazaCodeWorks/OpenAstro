```python
from astropy.io import fits
```

```python
datadir='/home/jp/Downloads/'
```

```python
kband=fits.getdata(datadir+'Kband_PI_mJy.fits')
hband=fits.getdata(datadir+'Hband_PI_mJy.fits')
```

```python
from astropy.visualization import ZScaleInterval
import matplotlib.pyplot as pl
%matplotlib inline

interval = ZScaleInterval()

vmin, vmax= interval.get_limits(kband)

xcenter,ycenter = (511,511)

pl.imshow(kband,#[xcenter-50:xcenter+50,ycenter-50:ycenter+50],
          interpolation=None,
          origin='lower',
          vmin=0,vmax=15,
          cmap='viridis',
          )

pl.colorbar()
```

```python
import pandas as pd

lr_K=pd.read_csv(datadir+'spiral_lr_K.dat',delimiter=' ',dtype=int)
lr_K.columns = ['x','y','fill']

ul_K=pd.read_csv(datadir+'spiral_ul_K.dat',delimiter=' ',dtype=int)
ul_K.columns = ['x','y','fill']
```
