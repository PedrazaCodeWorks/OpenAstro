## Everest pipeline lc

```python
epics='211314705,211579112,211763214,211965883,212088059,212132195,211314705,\
211579112,211763214,211965883,212088059,212132195,212278644,212297394,\
212330265,212398508,212420823,212440430,212480208,212495601,212496592,\
212543933,212563850,212572439,212737443'.split(',')

from tqdm import tqdm

everest={} #stellar params
for epic in tqdm(epics):
    try:
        star = k2plr.EVEREST(int(epic))
        everest[epic]=star
    except Exception as e:
        print(e)
```

Remove outliers

```python
import matplotlib.pyplot as pl
%matplotlib inline
import pandas as pd
import numpy as np

f=everest[epic].flux
t=everest[epic].time

sigma = 1.5

df=pd.DataFrame(np.c_[t,f],columns=['t','f'])
#retain values within sigma from the mean
mask = np.abs(df-df.mean())<=(sigma*df.std())
df = df[mask]

pl.plot(df.t,df.f)
```

```python
import pandas as pd
from astropy.stats import sigma_clip
import numpy as np

sigma = 1.5

for k in sorted(everest.keys()):
    f=everest[k].flux
    t=everest[k].time
    df=pd.DataFrame(np.c_[t,f],columns=['t','f'])
    #retain values within sigma from the mean
    mask = np.abs(df-df.mean())<=(sigma*df.std())
    df = df[mask]
    #remove outliers
    everest[k] = df
    print('{}:\nno. of outliers {}'.format(k, np.sum(~mask['f'])))
```

```python
import matplotlib.pyplot as pl
%matplotlib inline

ncols = 2
nrows = int(len(everest.keys())/ncols)+1

fig, ax = pl.subplots(nrows, ncols, figsize=(15,35))
for n,epic in enumerate(sorted(everest.keys())):
    f=everest[epic].f
    t=everest[epic].t
    if ncols == 1:
        ax[n].plot(t,f)
        ax[n].set_title(epic)
    else:
        ax[int(n/ncols),n%ncols].plot(t,f)
        ax[int(n/ncols),n%ncols].set_title(epic)
fig.tight_layout()
```
