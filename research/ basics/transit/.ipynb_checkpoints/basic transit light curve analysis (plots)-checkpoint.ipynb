## Stellar density
* asumming spherical star,
* Seager & Mallen-Ornelas (2003)
$$
\rho_*+k^3\rho_p=\frac{3\pi}{GP^2}\Big(\frac{a}{R_*} \Big)^3
$$
This can be derived from Kepler’s third law (Seager &
Mallen-Ornelas 2003). Since $k^3$ is usually small, the second
term on the left side of equation (30) is often negligible
and ρ⋆ can be determined purely from transit photometry.
This method for estimating ρ⋆ has proven to be a useful
diagnostic in photometric transit surveys: a true transit signal
should yield a value of ρ⋆ that is consistent with expectations
for a star of the given luminosity and spectral type

```python
def rho_star():
    return 3*np.pi*a**3/(G*P**2*Rs**3)
```

```python
def t_transit(a,Rs,Ms):
    return 2*Rs*np.sqrt(a/(G*Ms))
```

star-planet distance (Murray and Correia)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def star_planet_distance(a, e, f):
    '''
    a: semimajor axis of the relative orbit
    f: true anomaly (depends on e and p)
    e: eccentricity
    '''
    r=a*(1-e**2)/(1+e*np.cos(f))
    return r
```

eclipses occur when $r_{sky}$ is a local minimum:

```python
def r_sky(r,w,f,inc):
    X= -r*np.cos(w+f)
    Y= -r*np.sin(w+f)*np.cos(inc)
    Z= r*np.sin(w+f)*np.sin(inc)
    #r*np.sqrt(1-np.sin(w+f)**2*sin(i)**2)
    return np.sqrt(X**2+Y**2)
```

```python
#assumptions: e & w known from Doppler
a=1
e=0.1
f=np.radians(90)

w=1
inc=np.radians(90)

for j in np.linspace(0,180,18):
    f = np.radians(j)
    r=star_planet_distance(a,e,f)
    plt.plot(j,r_sky(r,w,j,inc), 'ko')
```

```python
def impact_param(a, inc, e, w, R_s):
    return a*np.cos(inc)/R_s*((1-e**2)/(1+e*np.sin(w)))
```

transit probability is calculated as the shadowed fraction of the line of longitude:

```python
def p_trans(R_p, R_s, w, e, a, grazing=True, approx=False):
    '''
    np.abs(b)<1+k
    '''
    if grazing:
        return (R_p+R_s)/a*(1+np.sin(w))/(1-e**2)
    elif approx:
        '''
        when R_p << R_s & e=0
        '''
        return R_s/a
    else:
        #minus
        return (R_p-R_s)/a*(1+np.sin(w))/(1-e**2)
```

## transit model

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib as mpl
mpl.style.use('seaborn-white')

mpl.rcParams.update({'font.size': 22})
#mpl.rcParams.update({'lines.linewidth': 5})
```

```python
from pytransit import MandelAgol
MA = MandelAgol()

def model_u(theta, t, p):
    k,tc,a,i,u1,u2 = theta
    m = MA.evaluate(t, k, (u1, u2), tc, p, a, i)
    return m
```

Assuming the ff transit parameters:
* period, $p$
* transit duration, $t_{14}$
* $k = Rp/R*$
* transit center, $t_c$
* scaled semi-major axis, $a_s$
* limb darkening coeffs, $u_1,u_2$
* inclination, $i$
* impact parameter, $b$

```python
def scaled_a(p, t14, k, inc=np.pi/2, b=0):
    numer = np.sqrt( (k + 1)**2 - b**2 )
    denom = np.sin(inc) * np.sin(t14 * np.pi / p)
    return float(numer / denom)
```

effect of changing radius ratio, k

```python
p_ = 5.0 
t14_ = 0.1

ks = np.linspace(0,1,10)

for k in ks:
    a_s = scaled_a(p_, t14_, k)
    plt.plot(k, a_s, 'o')
```

effect of impact parameter, b

```python
p_ = 5.0 
t14_ = 0.1
k_ = 0.1

bs = np.linspace(0,1,10)

for b in bs:
    a_s = scaled_a(p_, t14_, k_, b=b)
    plt.plot(b, a_s, 'o')
```

```python
k_,tc_,t14_ ,inc_,u1_,u2_ = 0.1, 0, 0.5, np.radians(90), 0.4, 0.4

a_s = scaled_a(p_, t14_, k_, inc_, b=0)
theta_  = [k_,tc_,a_s,inc_,u1_,u2_]

t = np.linspace(-1, 1, 1000)
p_= 8.0 
model= model_u(theta_, t, p_)

plt.plot(t, model, '-')
plt.xlabel('Phase [days]')
plt.ylabel('Relative Flux')
```

```python
mpl.rcParams.update({'font.size': 16})

k_   = 0.1
tc_  = 0.0
t14_ = 0.5
inc_ = np.radians(90)
u1_,u2_ = 0.4, 0.4

t = np.linspace(-0.5, 0.5, 1000)
p_= 8.0 

fig, ax = plt.subplots(3,2,figsize=(15,20))
ks = np.linspace(0.01,0.5,10)
for j in ks:
    #k
    a_s1 = scaled_a(p_, t14_, j, inc_, b=0)
    theta1_=[j,tc_,a_s1,inc_,u1_,u2_]
    model1= model_u(theta1_, t, p_)
    l='{:.2f}'.format(j)
    ax[0,0].plot(t, model1, '-',label=l)
    
    #t14
    a_s2=scaled_a(p_, j, k_, inc_, b=0)
    theta2_=[k_,tc_,a_s2,inc_,u1_,u2_]
    model2= model_u(theta2_, t, p_)
    ax[0,1].plot(t, model2, '-', label=l)
    
    #u1
    a_s3=scaled_a(p_, t14_, k_, inc_, b=0)
    theta3_=[k_,tc_,a_s3,inc_,j,u2_]
    model3= model_u(theta3_, t, p_)
    ax[1,0].plot(t, model3, '-', label=l)
    
    #u2
    a_s4=scaled_a(p_, t14_, k_, inc_, b=0)
    theta4_=[k_,tc_,a_s4,inc_,u1_,j]
    model4= model_u(theta4_, t, p_)
    ax[1,1].plot(t, model4, '-', label=l)
    
    #impact param
    a_s5=scaled_a(p_, t14_, k_, inc_, b=j)
    theta5_=[k_,tc_,a_s5,inc_,u1_,j]
    model5= model_u(theta5_, t, p_)
    ax[2,0].plot(t, model5, '-', label=l)
    
#50 to 90 deg
incls=np.linspace(np.radians(50), np.radians(90), 10)
for incl in incls:
    #inc
    a_s6=scaled_a(p_, t14_, k_, incl, b=0)
    theta6_=[k_,tc_,a_s6,inc_,u1_,u2_]
    model6= model_u(theta6_, t, p_)
    l='{:.2f}'.format(np.degrees(incl))
    ax[2,1].plot(t, model6, '-', label=l)
    
    
ax[0,0].set_xlabel('Phase [days]')
ax[0,0].set_ylabel('Relative Flux')
ax[0,0].legend(title='k', loc='best')
ax[0,1].legend(title=r'$t_{14}$', loc='best')
ax[1,0].legend(title=r'$u_1$', loc='best')
ax[1,1].legend(title=r'$u_2$', loc='best')
ax[2,0].legend(title=r'$b$', loc='best')
ax[2,1].legend(title=r'$i$', loc='best')
```

* t14 and inc have qualitatively similar effect: large --> broader light curve
* larger u1/u2 --> deeper

## Fitting a light curve: $\chi^2$

```python
#chi-squared minimization
def obj(theta, t, p, f):
    m = model_u(theta, t, p)
    return np.sum((m-f)**2)
```

```python
import scipy.optimize as op

k_   = 0.1
tc_  = 0.0
t14_ = 0.5
inc_ = np.radians(90)
u1_,u2_ = 0.4, 0.4

t = np.linspace(-0.5, 0.5, 1000)
p_= 8.0

a_s_ = scaled_a(p_, t14_, k_, inc=inc_, b=0)
theta_true = [k_,tc_,a_s_,inc_,u1_,u2_]
f = model_u(theta_true, t, p_)

#initial guesses
p0    =9.0
t14_0 =0.6
k0    =0.2
tc0   =0.1
inc0  =np.radians(90)
u1_0,u2_0 = 0.5,0.5

a_s0 = scaled_a(p0, t14_0, k0, inc=inc0, b=0)
theta0 = [k0,tc0,a_s0,inc0,u1_,u2_]

res = op.minimize(obj, theta0, args=(t, p0, f), method='nelder-mead')
```

```python
#best fit parameters
res.x
```

```python
plt.plot(t, model_u(theta_true, t, p_), 'o', label='data', alpha=0.8)
plt.plot(t, model_u(res.x, t, p0), '-', label='fit', lw=5)
plt.legend()
plt.xlabel('Phase [days]')
plt.ylabel('Relative Flux')
```

## Sys-Rem

* [Tamuz (2005)](https://arxiv.org/pdf/astro-ph/0502056.pdf)
* c.f. [Kruszewski & Semeniuk (2003)]()
* based on lower rank approximation of matrices
* general algorithm to deal with any linear systematic effects
* reduces to Principal Component Analysis (Murtagh & Heck 1987, Ch. 2), when different uncertainties of the data points ignored

* Color-dependent atmospheric extinction is an obvious observational effect that contaminates ground-based photometric measurements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import getpass
from tqdm import tqdm
%matplotlib inline
data_dir = '/home/'+getpass.getuser()+'/data/transit/hatp44_data/lc_hatp44_msct_170215'
file_list=glob.glob(data_dir+'/*.dat')
file_list.sort()

name='hatp44'
date='170215'
target_star_id='2'
comparison_star_id='1'#13, #3
radii_range='9-14'

data={}
data2={} #outliers remoed
bands='g,r,z'.split(',')
for b in bands:
    fname='lcf_msct_'+b+'_'+name+'_'+date+'_t'+target_star_id+'_c'+comparison_star_id+'_r'+radii_range+'.bjd.dat'
    df=pd.read_csv(os.path.join(data_dir,fname), delimiter=' ', parse_dates=True)
    df = df.set_index('BJD(TDB)-2450000')
    try:
        df=df.drop('Unnamed: 20',1)
    except:
        pass
    #df.head()
    data[b]=df
```

```python
def model_s(theta, fwhm, xcenter, ycenter, t):
    #functional form of systematics model
    offset = np.ones(len(t))
    s = (np.array(theta)*np.c_[fwhm, xcenter, ycenter, offset, t]).sum(axis=1)
    #s = [np.sum(c[k] * x**k) for k in np.arange(N)]
    # unpack 6 free params
    #a,b,c,d,e = theta
    #s = a + b*t + c*fwhm + d*xcenter + e*ycenter
    return s
```
