The timeseries is stored in `star.time` and `star.flux` (PLD-de-trended flux) or `star.fcor` (de-trended flux with CBV correction). The indices of all outliers are stored in `star.mask`.

* pixel level decorrelation (PLD)
* co-trending basis vector (CVB)

```python
import pandas as pd

#convert to df
df1 = pd.DataFrame(EPIC153.flux, columns=['EPIC153'], index=EPIC153.time)
df2 = pd.DataFrame(EPIC317.flux, columns=['EPIC317'], index=EPIC317.time)
df3 = pd.DataFrame(EPIC916.flux, columns=['EPIC916'], index=EPIC916.time)
df4 = pd.DataFrame(EPIC548.flux, columns=['EPIC548'], index=EPIC548.time)
df5 = pd.DataFrame(EPIC637.flux, columns=['EPIC637'], index=EPIC637.time)
df6 = pd.DataFrame(EPIC363.flux, columns=['EPIC363'], index=EPIC363.time)
df7 = pd.DataFrame(EPIC512.flux, columns=['EPIC512'], index=EPIC512.time)
df8 = pd.DataFrame(EPIC387.flux, columns=['EPIC387'], index=EPIC387.time)
df9 = pd.DataFrame(EPIC911.flux, columns=['EPIC911'], index=EPIC911.time)
```

```python
#save to csv
df1.to_csv('data/203843911.csv')
df2.to_csv('data/203824153.csv')
df3.to_csv('data/203937317.csv')
df4.to_csv('data/204278916.csv')
df5.to_csv('data/204469637.csv')
df6.to_csv('data/204514548.csv')
df7.to_csv('data/204630363.csv')
df8.to_csv('data/204638512.csv')
df9.to_csv('data/205151387.csv')
```

## Phase-folding 

If there are transits/eclipses in a light curve, everest can use the GP prediction to whiten the timeseries and fold it on the period of the planet. If the time of first transit and period of an exoplanet/EB are known, plotting the folded transit/eclipse is easy. Just remember to mask the transit and re-compute the model beforehand:

```python
EPIC911.time == EPIC363.time
```

EPIC916.mask_planet(1980.42, 10.054)
EPIC916.compute()
EPIC916.plot_folded(1980.42, 10.054)
