```python
from astropy import units as u
from tqdm import tqdm

targets = "204514548,203410665,204864076,204638512,204630363,205151387,203843911,203937317,204278916,204512343,203950167,204469637,205519771,204932990,203824153,204137184,204530046,203995761,204472612,203343161,204211116,205241182,204329690,203770366,203850058".split(',')

ra, dec, kpmag, jmag, hmag, kmag, rmag, teff, twomass, distance = [], [], [], [], [], [], [], [], [], []

for epic in tqdm(targets):
    star = client.k2_star(epic)
    ra.append(star.k2_ra)
    dec.append(star.k2_dec)
    jmag.append(star.kp)
    hmag.append(star.hmag)
    kmag.append(star.kmag)
    rmag.append(star.rmag)
    kpmag.append(star.kp)
    teff.append(star.teff)
    twomass.append(star.twomass)
    distance.append(star.d)
```

```python
ra
```

```python
ra_h=ra*u.degree
ra_h=ra_h.to(u.hourangle)
```

```python
ra_h.value
```

```python
import pandas as pd

df = pd.DataFrame({'EPIC': targets, 'RA (deg)':ra, 'RA (hr)':ra_h.value, 'DEC (deg)': dec, 'Kpmag':kpmag, 
                   'Jmag':jmag, 'Hmag':hmag, 'Kmag':kmag, 'Rmag':rmag, 'Teff':teff, '2MASS':twomass, 'distance': distance})
```

```python
df
```

```python
df.to_csv('ipynb_output.csv')#, delimiter='\tab')
```
